```python
# Define bins as numeric values
bins = [-float('inf'), 100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']


# Create the Income Band
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Define categorical columns as a list
categorical_cols = ['BRId', 'GenderId', 'Properties Owned', 'Risk Weighting', 'IAId', 
                    'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 
                    'Amount of Credit Cards']
# Iterate over the categorical columns and display value counts
for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

```python

# Define the list of categorical predictors
categorical_cols = ['BRId', 'GenderId', 'Properties Owned', 'Risk Weighting', 'IAId', 
                    'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 
                    'Amount of Credit Cards']

# Create a count plot for each predictor
for i, predictor in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
    plt.title(f'Count Plot of {predictor} by Gender')
    plt.xticks(rotation=45)  # Rotate x labels for better readability
    plt.tight_layout()  # Adjust layout to prevent clipping
    plt.show()  # Display the plot
```

```python
for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# Define the numerical columns
numerical_col = ['Age', 'Location ID', 'Estimated Income', 'Superannuation Savings', 
                 'Amount of Credit Cards', 'Credit Card Balance', 'Bank Loans', 
                 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 
                 'Foreign Currency Account', 'Business Lending', 'Properties Owned', 
                 'Risk Weighting', 'IAId']

# Define the number of rows and columns for the subplots
n_cols = 3  # Number of columns
n_rows = (len(numerical_col) + n_cols - 1) // n_cols  # Calculate number of rows needed

plt.figure(figsize=(15, 5 * n_rows))  # Adjust figure size based on number of rows

# Define a list of colors for each column
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf',
          '#ffbb78', '#98df8a', '#ff9896', '#c5b0d5', '#c49c94', 
          '#f7b6d2']

for i, col in enumerate(numerical_col):
    plt.subplot(n_rows, n_cols, i + 1)  # Create a new subplot
    sns.histplot(df[col], kde=True, color=colors[i % len(colors)])  # Set color for each histogram
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python

correlation_matrix = df[numerical_col].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot= True, cmap='Blues', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```
